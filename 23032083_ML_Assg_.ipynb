```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
import pandas as pd

df = pd.DataFrame(data.data, columns=data.feature_names)

# Add target and target class names as new columns
df['target'] = data.target
df['class'] = df['target'].apply(lambda x: data.target_names[x])

print(df)
```

```python
X, y = data.data, data.target
```

```python
from sklearn.tree import DecisionTreeClassifier

tr = DecisionTreeClassifier()
tr.fit(X, y)
```

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(50,30))
class_names=data.target_names.tolist()
plot = tree.plot_tree(tr, feature_names=data.feature_names, class_names=class_names, filled=True)
```

```python
print(tree.export_text(tr))
```
